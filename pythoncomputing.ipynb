```python
#1번 문항
#구구단 5단을 출력하는 프로그램 작성
#i를 1~9까지 반복
for i in range (1,10) :
    # 5단 계산
    multi = 5 * i
    #결과 출력
    print(f'5 x {i} = {multi}')
```

```python
#2번 문항
#반복문으로 1~5 까지 숫자를 1~5번씩 출력하는 프로그램 작성
#i를 1~5까지 반복
for i in range(1,6):
    #i를 문자로 만들어서 숫자 i만큼 출력
    print (str(i) * i)
```

```python
#3번 문항
#주이진 숫자와 연잔자에 대해 계산을 수행하는 프로그램 작성
#calculation 리스트
calculation = [
    (10, '*', 5),
    (20, '-', 8),
    (6, '*', 7),
    (15, '/', 3),
    (8, '/', 0), #0으로 나누기
    (12, '%', 5) #잘못된 연산자
]

print("<계산 결과>")
for num1, ope, num2 in calculation :
    if ope == '+' :
        result = num1 + num2
        print(f'{num1} {ope} {num2} = {result}')
    elif ope == '-' :
        result = num1 - num2
        print(f'{num1} {ope} {num2} = {result}')
    elif ope == '*' :
        result = num1 * num2
        print(f'{num1} {ope} {num2} = {result}')
    elif ope == '/' :
        if num2 == 0 :
            print("0으로 나눌 수 없습니다.")
        else :
            result = num1 / num2
            print(f'{num1} {ope} {num2} = {result}')
    else :
        print("잘못된 연산자 입니다.")
```
